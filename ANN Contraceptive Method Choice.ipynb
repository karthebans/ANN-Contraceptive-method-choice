```python
import pandas as pd
```

```python
df = pd.read_csv('cmc.data')
```

```python
df.columns = ['Wifes age', 'Wifes education', 'Husbands education', 'Number of children ever born',
                                 'Wifes religion', 'Wifes now working?', 'Husbands occupation', 'Standard-of-living index',
                                 'Media exposure', 'Contraceptive method used']
```

```python
df
```

```python
x = df.iloc[:,:-1].values
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.25, random_state=10)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.Input((9,)))
model.add(tf.keras.layers.Dense(100, activation = 'relu'))
model.add(tf.keras.layers.Dense(100, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation = 'softmax'))
```

```python
model.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 100, use_multiprocessing=True)
```
